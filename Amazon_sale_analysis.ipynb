## Group By

The groupby() function in pandas is used to group data based on one or more columns in a DataFrame

```python
S_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty,hue='Size')
```

##### Note: From above Graph you can see that most of the Qty buys M-Size in the sales

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(15,5))
plt.hist(column_data,bins=50,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()

```

##### Note: From above Graph you can see that most of the  buyers are T-shirt

```python
# Checking B2B Data  by using pie chart
B2B_check=df['B2B'].value_counts()
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
plt.show()
```

##### Note : From above chart  we can see that maximum i.e. 99.3% of buyers are retailers and 0.7% are B2B buyers

```python
#  Prepare data for pie chart & plot pie chart:
a1=df['Fulfilment'].value_counts()
plt.pie(a1,labels=a1.index,autopct='%1.1f%%',radius=1,wedgeprops=dict(width=0.6))
plt.show()

```

##### Note: From above chart you can see that most of the  Fulfilment  are from amazon

```python
import matplotlib.pyplot as plt

#Prepare data for scatter plot:
x_data=df['Category']
y_data=df['Size']

#plot a scatter plot:
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#Plot a count graph for state:
plt.figure(figsize=(25,20))
sns.countplot(x='ship-state', data=df, hue='ship-state')
plt.xlabel('state')
plt.ylabel('count')
plt.xticks(rotation=90)
plt.show()
```

```python
#plot countplot for top 10 'ship-state':
top_10_states=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(7,5))
sns.countplot(x='ship-state', data=df, palette='Set2', order=top_10_states.index,hue='ship-state',legend=False)
plt.xlabel('state')
plt.ylabel('count')
plt.title('Distribution By Top 10 States')
plt.xticks(rotation=45)
plt.show()
```

#### Note: From above Graph you can see that most of the  buyers are from Maharashtra state
